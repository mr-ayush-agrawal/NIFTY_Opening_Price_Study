# PreProcessing the data of Different Stock Market

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Aord = pd.read_csv('./Indices/ALLOrdinary.csv',index_col=0)             # Opening Time(IST) -> 5:30 am
Nikkei = pd.read_csv('./Indices/Nikkei225.csv',index_col=0)             # Opening Time(IST) -> 6 Am
Hsi = pd.read_csv('./Indices/HSI.csv',index_col=0)                      # Opening Time(IST) -> 7 am
Cac40 = pd.read_csv('./Indices/CAC40.csv',index_col=0)                  # Opening Time(IST) -> 12:30 pm
Daxi = pd.read_csv('./Indices/DAXI.csv',index_col=0)                    # Opening Time(IST) -> 12:30 pm
Dji = pd.read_csv('./Indices/DJI.csv',index_col=0)                      # Opening Time(IST) -> 7 pm
Nasdaq = pd.read_csv('./Indices/nasdaq_composite.csv',index_col=0)      # Opening Time(IST) -> 7 pm
Sp500 = pd.read_csv('./Indices/SP500.csv',index_col=0)                  # Opening Time(IST) -> 7 pm
Spy = pd.read_csv('./Indices/SPY.csv',index_col=0)                      # Opening Time(IST) -> 7 pm

NIFTY = pd.read_csv('./Indices/NIFTY.csv',index_col=0)                  # Opening time(IST) -> 9:15aM
```

We will be using the same day's data for the market which opens before the Indian Stock Market NSE's **NIFTY** and for those which open late we shall be using the opening Price for the day Before.

```python
NIFTY.head(10)
```

## Taking Opening prices of all the data

```python
All_Indices = pd.DataFrame(index=NIFTY.index)

All_Indices['Aord'] = Aord['Open']
All_Indices['Nikkei'] = Nikkei['Open']
All_Indices['Hsi'] = Hsi['Open']
All_Indices['Cac40'] = Cac40['Open']
All_Indices['Daxi'] = Daxi['Open']
All_Indices['Dji'] = Dji['Open']
All_Indices['Nasdaq'] = Nasdaq['Open']
All_Indices['Sp500'] = Sp500['Open']
All_Indices['Spy'] = Spy['Open']

All_Indices['NIFTY'] = NIFTY['Open']

All_Indices
```

```python
All_Indices.head(10)
```

```python
All_Indices.tail(10)
```

```python
All_Indices.fillna(method='ffill',inplace=True)
```

```python
All_Indices.isnull().sum()
```

```python
All_Indices.index = pd.DatetimeIndex(All_Indices.index)
type(All_Indices.index[0])
```

### Shify Index shows what was the last opening prices of the Different stock indices before the opening of NIFTY

```python
Shift_index = pd.DataFrame(index=All_Indices.index)

Shift_index['Aord'] = All_Indices['Aord']
Shift_index['Nikkei'] = All_Indices['Nikkei']
Shift_index['Hsi'] = All_Indices['Hsi']
Shift_index['Daxi'] = All_Indices['Daxi'].shift(1)
Shift_index['Dji'] = All_Indices['Dji'].shift(1)
Shift_index['Nasdaq'] = All_Indices['Nasdaq'].shift(1)
Shift_index['Sp500'] = All_Indices['Sp500'].shift(1)
Shift_index['Spy'] = All_Indices['Spy'].shift(1)

Shift_index['NIFTY_Previous'] = All_Indices['NIFTY'].shift(1)
Shift_index['NIFTY'] = All_Indices['NIFTY']
```

```python
Shift_index.head(15)
```

```python
Shift_index.drop('2018-08-23',axis=0,inplace=True)
Shift_index.dropna(inplace=True)
Shift_index
```

```python
Shift_index.isnull().sum()
```

```python
Shift_index.to_csv("./Indices/Processed_Indices.csv")
```

```python
plt.figure(figsize=(20,12))
for x in Shift_index.columns:
    if x == 'NIFTY_Previous':
        continue
    plt.plot(Shift_index.index,Shift_index[x],label=str(x))
plt.xlabel("Time Period")
plt.ylabel("Points")
plt.grid(True)
plt.legend()
```

```python
plt.figure(figsize=(13,8))
sns.heatmap(Shift_index.drop(columns='NIFTY_Previous').corr(),cmap='Greens',annot=True)
```
